```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = 'dataset/train.csv'
df = pd.read_csv(data_path)
```

```python
df_sex = df.Sex
```

```python
sx_dummies = pd.get_dummies(df_sex, prefix='sex')  #prefix : 컬럼 앞에다가 고정적으로 붙여줌
```

```python
sx_dropped = df.drop('Sex',axis=1) #Sex 컬럼 없애줌
new_df = sx_dropped.join(sx_dummies)
new_df.head()
```

```python
df_embark = df.Embarked
em_dummies = pd.get_dummies(df_embark,prefix='embark')
em_dropped = new_df.drop('Embarked',axis=1)
new_df = em_dropped.join(em_dummies*3)
new_df.head()
```

```python
new_df = new_df.drop(['PassengerId','Name','Ticket','Cabin'],1)
new_df.head()
```

```python
df.head()
```

```python
org_df = df.drop(['PassengerId','Name','Sex','Age','Ticket','Cabin','Embarked'],1)
#org_df = new_df.dropna()#fillna(value=-1)

num_dataset = int(0.8 * len(org_df))

trainset = org_df[:num_dataset]
testset  = org_df[num_dataset:]

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(trainset[trainset.columns[1:]], trainset.Survived)

preds = model.predict(testset[trainset.columns[1:]])
np.mean(preds == testset.Survived)
```

```python
org_df = new_df.dropna()#fillna(value=-1)

num_dataset = int(0.8 * len(org_df))

trainset = org_df[:num_dataset]
testset  = org_df[num_dataset:]

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(trainset[trainset.columns[1:]], trainset.Survived)

preds = model.predict(testset[trainset.columns[1:]])
np.mean(preds == testset.Survived)
```
